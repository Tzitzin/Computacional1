```python
#LEER ARCHIVO Y DAR NOMBRE A COLUMNAS
df = pd.read_csv("Act_7_bermudas.csv", names=['WaterLevel', 'Hora'],skiprows=[0])

#Mostrar c√°lculos estadisticos de los datos. 
df.describe()
```

```python
%matplotlib inline

#Tama√±o de la imagen
fig = plt.gcf()
fig.set_size_inches(13, 8)

#Ejes
x = df[u'Hora']
y = df[u'WaterLevel']

#Graficar
mplt.plot(x,y)

#Etiquetas
mplt.ylabel('Water Level(m)')
mplt.xlabel('N√∫mero de hora')
mplt.title('Nivel del mar [Enero(2017)-Febrero(2017),Bermuda')

#Malla
plt.grid()
```

```python
def f(t):
 return 0.4+ (0.04)*np.cos(np.pi*t/13.16) + (0.07)*np.cos(np.pi*t/12.19) + (0.06)*np.cos(np.pi*t/6.3694) + (0.37)*np.cos(np.pi*t/6.25) +(0.07)*np.cos(np.pi*t/6.06)

#interpolar la funcion
df[u'WaterLevel'] = df[u'WaterLevel'].astype(float).interpolate(method='spline', order=2)

fig = mplt.gcf()
fig.set_size_inches(12, 7)

mplt.plot(df[u'Hora'], f(df[u'Hora']), c='green', label='Curva construida')
mplt.plot(df[u'Hora'], df[u'WaterLevel'], c='blue', label='Curva original')
#plt.xlim(pd.Timestamp('2017-01-04 00:00:00'), pd.Timestamp('2017-01-31 00:23:00'))


mplt.ylabel('Nivel de la marea(m)')
mplt.xlabel('Hora')
mplt.legend(fancybox=True, shadow=True)
mplt.grid(True)
```
