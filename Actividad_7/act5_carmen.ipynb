```python
import pandas as pd
import numpy as np
import matplotlib as plt
import statsmodels.api as sm
import scipy.stats as stats
from pylab import *
from scipy.io import loadmat
```

```python
df=pd.read_csv("carmenenero.csv", header=int(0))
```

```python
df.head()
```

```python
from datetime import datetime
df['date']= df.apply(lambda x:datetime.strptime("{0} {1} {2} {3}".format(x[u'anio'],x[u'mes'], x[u'dia'], x[u'hora(utc)']), "%Y %m %d %H"),axis=1)
```

```python
df.head()
```

```python
x=df["date"]
y=df["altura(mm)"]

plt.title("Mareas en Ciudad del Carmen, Enero 2016 ")
plt.ylabel('Nivel del agua (mm)')
plt.xlabel('Fecha')
plt.grid(True)


plt.plot(x,y)
plt.show()
```
